```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df_import = pd.read_csv('/Users/janmikolon/Wine Files/drive_c/users/janmikolon/Desktop/FV-L100-7-7-P1/Results/Struktur/matrix_new_filtered.csv')

df_import=df_import.drop(['Unnamed: 0'],axis=1)

#print(df_import)

df1=df_import.applymap(lambda x: '___' in str(x))
df2 = df_import[df1.any(axis=1)]
df2=df2.drop_duplicates(keep=False)
colN = df2.columns.values[:17]
for q in colN:
    #print(q)
    q= str(q)
    for index, row in df2.iterrows():
        meine_suche= str(row[q])

        if meine_suche.startswith('___'):
                    row[q] = "Startkante"
        elif "___" in meine_suche:
                splitten= meine_suche.split("___")
                if splitten[0]==splitten[1]: 
                    row[q] = "Haltekante"
                elif splitten[0] !=splitten[1]:
                    row[q] ="Fahrkante"
                    
                    
                    
                    
df2
```
